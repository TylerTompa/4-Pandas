# In this notebook, we will read csv files containing data about various schools and students thereof, and create dataframes therefrom in order to help the school board and mayor make strategic decisions regarding future school budgets and priorities.

```python
# We will import the pandas package, so that we can read csv files, create dataframes therefrom, and perform data analysis.
# We will also import numpy to assist in mathematical computations.
import pandas as pd
import numpy as np

pd.options.display.float_format = "{:,}".format
```

```python
# This creates file paths to our school data and student data csv files, so that we can later create dataframes therefrom.
schools_complete_file_path = "Resources/schools_complete.csv"
students_complete_file_path = "Resources/students_complete.csv"
```

```python
# This creates dataframes from the csv files we read earlier.
schools_dataframe = pd.read_csv(schools_complete_file_path)
students_dataframe = pd.read_csv(students_complete_file_path)
```

```python
schools_dataframe.head()
```

```python
students_dataframe.head()
```

```python
# This merges the students dataframe with the schools dataframe by the common field "school_name"
students_schools_merged_dataframe = pd.merge(students_dataframe, schools_dataframe, how="left", on=["school_name", "school_name"])
students_schools_merged_dataframe.head()
```

```python
# By using the .nunique() method, we count the number of unique values in the school_name column; this is the total number of schools.
total_schools = students_schools_merged_dataframe["school_name"].nunique()

# By using the .count() method, we count the number of fields in the student_name column; this is the total number of students.
total_students = students_schools_merged_dataframe["student_name"].count()

# By using the .sum() method, we calculate the sum of each school budget.  We must first use the .unique() method, to ensure that we do not add the budget from the same school more than once.
total_budget = students_schools_merged_dataframe["budget"].unique().sum()

# By using the .mean method, we calculate the average math score of every student.
average_math = students_schools_merged_dataframe["math_score"].mean()

# By using the .mean method, we calculate the average math score of every student.
average_math = students_schools_merged_dataframe["math_score"].mean()

# By using the .mean method, we calculate the average reading score of every student.
average_reading = students_schools_merged_dataframe["reading_score"].mean()

# By using the function average from the numpy package, we can calculate the average of the values written therein.
overall_passing_rate = np.mean([average_math, average_reading])

# We ust a pandas feature that allows us to count the number of values in a column based on a condition- in our case, grades greater than or equal to 70- then we use the len() function to get the length of this list.
# We then divide this number by the length of the entire list, and multiply of 100, giving us the percent of passing grades.
percent_passing_math = len(students_schools_merged_dataframe[students_schools_merged_dataframe["math_score"] >= 70]) / students_schools_merged_dataframe["math_score"].count() * 100

# We ust a pandas feature that allows us to count the number of values in a column based on a condition- in our case, grades greater than or equal to 70- then we use the len() function to get the length of this list.
# We then divide this number by the length of the entire list, and multiply of 100, giving us the percent of passing grades.
percent_passing_reading = len(students_schools_merged_dataframe[students_schools_merged_dataframe["reading_score"] >= 70]) / students_schools_merged_dataframe["reading_score"].count() * 100
```

```python
# This creates a dictionary within a list, which we will later be able to use as the data for the single-record district_summary dataframe.
district_summary_data = [{"Total Schools":total_schools,
                          "Total Students":total_students,
                          "Total Budget":total_budget,
                         "Average Math Score":average_math,
                         "Average Reading Score":average_reading,
                         "Percent Overall Passing Rate":overall_passing_rate,
                         "Percent Passing Math":percent_passing_math,
                         "Percent Passing Reading":percent_passing_reading}]
```

```python
# This creates the single-record district summary dataframe using the data we defined earlier.
district_summary = pd.DataFrame(data=district_summary_data, columns=["Total Schools",
                                                                     "Total Students",
                                                                    "Total Budget",
                                                                    "Average Math Score",
                                                                    "Average Reading Score",
                                                                    "Percent Overall Passing Rate",
                                                                    "Percent Passing Math",
                                                                    "Percent Passing Reading"])
district_summary
```

## In this section, we create a dataframe that summarizes data per school.

```python
# This creates lists of school names, types, total students, and budgets, which we will later set as the fields in our schools_summary dataframe.
school_names = [name for name in schools_dataframe["school_name"]]
school_types = [type for type in schools_dataframe["type"]]
total_students = [size for size in schools_dataframe["size"]]
budgets = [budget for budget in schools_dataframe["budget"]]

# This creates series of average math and reading score for each school.
average_math_scores = students_dataframe.groupby("school_name", sort=False)["math_score"].mean()
average_reading_scores = students_dataframe.groupby("school_name", sort=False)["reading_score"].mean()

# This creates series of percent passing math and reading for each school.
percent_passing_math_by_school = students_dataframe[(students_dataframe["math_score"] >= 70)].groupby("school_name", sort=False)["math_score"].count() / students_dataframe.groupby("school_name", sort=False)["math_score"].count() * 100
percent_passing_reading_by_school = students_dataframe[(students_dataframe["reading_score"] >= 70)].groupby("school_name", sort=False)["reading_score"].count() / students_dataframe.groupby("school_name", sort=False)["math_score"].count() * 100
```

```python
# This creates the school_summary dataframe, then sets the fields thereof to the lists created just before this step.
school_summary = pd.DataFrame()
school_summary["School Name"] = school_names
school_summary["School Type"] = school_types
school_summary["Total Students"] = total_students
school_summary["Total Budget"] = budgets

school_summary["Per Student Budget"] = school_summary["Total Budget"] / school_summary["Total Students"]

# We must convert the average math and reading scores fields to lists to use the data therein in a dataframe.
school_summary["Average Math Score"] = average_math_scores.tolist()
school_summary["Average Reading Score"] = average_reading_scores.tolist()

# We must convert the percent passing math and reading fields to lists to use the data therein in a dataframe.
school_summary["Percent Passing Math"] = percent_passing_math_by_school.tolist()
school_summary["Percent Passing Reading"] = percent_passing_reading_by_school.tolist()

school_summary["Overall Passing Rate"] = school_summary[["Percent Passing Math", "Percent Passing Reading"]].mean(axis=1)

# Once we have all the data in our school_summary dataframe, we set the index to School Name.
school_summary = school_summary.set_index(["School Name"])

school_summary.head()
```

```python
# This uses the .nlargest method to shows us the top x amount of schools by the Overall Passing Rate, where we define x as 5.
school_summary.nlargest(5, "Overall Passing Rate", keep="first")
```

```python
# This uses the .nsmallest method to shows us the bottom x amount of schools by the Overall Passing Rate, where we define x as 5.
school_summary.nsmallest(5, "Overall Passing Rate", keep="first")
```

## In this section create dataframes which summarize various scores by school spending, school size, and school type.

```python
# First we view the school_summary dataframe to see the column names.
school_summary.head()
```

```python
# Now we view the minimum and maximum per student budget, in order to determine appropriate bins.
print(school_summary["Per Student Budget"].min())
print(school_summary["Per Student Budget"].max())
```

```python
# Based on our minimum and maximum per-student-budget, we define enough bins to make four groups of budgets.
budget_bins = [0, 585, 615, 645, 675]
budget_bins_group_names = ["<$585", "$585-615", "$615-645", ">$645"]
```

```python
# We will drop all irrelevant columns before moving on.
school_summary_by_per_student_budget = school_summary.drop(columns=["School Type", "Total Students", "Total Budget", "Per Student Budget"])
school_summary_by_per_student_budget
```

```python
# Here we make a new dataframe composed of the data in the school_summary dataframe as well as a column of per-student-budgets categories.
school_summary_by_per_student_budget["Spending Range (Per Student)"] = pd.cut(school_summary["Per Student Budget"], budget_bins, labels=budget_bins_group_names, include_lowest=False)
school_summary_by_per_student_budget.head()
```

```python
# The following lines create series of average scores based on conditions of per-student-budget categories.
spending_bin_1_averages = school_summary_by_per_student_budget[(school_summary_by_per_student_budget["Spending Range (Per Student)"] == "<$585")].mean()
spending_bin_2_averages = school_summary_by_per_student_budget[(school_summary_by_per_student_budget["Spending Range (Per Student)"] == "$585-615")].mean()
spending_bin_3_averages = school_summary_by_per_student_budget[(school_summary_by_per_student_budget["Spending Range (Per Student)"] == "$615-645")].mean()
spending_bin_4_averages = school_summary_by_per_student_budget[(school_summary_by_per_student_budget["Spending Range (Per Student)"] == ">$645")].mean()
```

```python
print(spending_bin_1_averages)
print(spending_bin_2_averages)
print(spending_bin_3_averages)
print(spending_bin_4_averages)
```

```python
# This creates a new dataframe using the series which contain the average scores based on per-student budget.
average_scores_by_per_student_budget = pd.DataFrame({"<$585": spending_bin_1_averages,
                        "$585-615": spending_bin_2_averages,
                        "$615-645": spending_bin_3_averages,
                        ">$645": spending_bin_4_averages})
average_scores_by_per_student_budget
```

```python
# Here we transpose the dataframe so that each per-student-budget makes up a row, and the scores make up the columns.
# We also name our index, making our dataframe easier to understand.
average_scores_by_per_student_budget = average_scores_by_per_student_budget.transpose()
average_scores_by_per_student_budget = average_scores_by_per_student_budget.rename_axis("Spending Ranges (Per Student)")
average_scores_by_per_student_budget
```

### In this section we will create a dataframe which summarizes scores by school size.

```python
# First we view the school_summary dataframe to see the column names.
school_summary.head()
```

```python
# Now we view the minimum and maximum values for total students, in order to determine appropriate bins.
print(school_summary["Total Students"].min())
print(school_summary["Total Students"].max())
```

```python
# Based on our minimum and maximum per-student-budget, we define enough bins to make four groups of budgets.
total_students_bins = [0, 1000, 2000, 5000]
total_students_bins_group_names = ["Small (<1000)", "Medium (1000-2000)", "Large (>2000)"]
```

```python
# We will drop all irrelevant columns before moving on.
school_summary_by_size = school_summary.drop(columns=["School Type", "Total Students", "Total Budget", "Per Student Budget"])
school_summary_by_size.head()
```

```python
# Here we make a new dataframe composed of the data in the school_summary dataframe as well as a column of school-size categories.
school_summary_by_size["School Size"] = pd.cut(school_summary["Total Students"], total_students_bins, labels=total_students_bins_group_names, include_lowest=False)
school_summary_by_size.head()
```

```python
# The following lines create series of average scores based on conditions of school size categories.
total_students_bin_1_averages = school_summary_by_size[(school_summary_by_size["School Size"] == "Small (<1000)")].mean()
total_students_bin_2_averages = school_summary_by_size[(school_summary_by_size["School Size"] == "Medium (1000-2000)")].mean()
total_students_bin_3_averages = school_summary_by_size[(school_summary_by_size["School Size"] == "Large (>2000)")].mean()
```

```python
print(total_students_bin_1_averages)
print(total_students_bin_2_averages)
print(total_students_bin_3_averages)
```

```python
# This creates a new dataframe using the series which contain the average scores based on school size.
average_scores_by_school_size = pd.DataFrame({"Small (<1000)": total_students_bin_1_averages,
                        "Medium (1000-2000)": total_students_bin_2_averages,
                        "Large (>2000)": total_students_bin_3_averages})
average_scores_by_school_size
```

```python
# Here we transpose the dataframe so that each school-size categories make up the rows, and the scores make up the columns.
# We also name our index, making our dataframe easier to understand.
average_scores_by_school_size = average_scores_by_school_size.transpose()
average_scores_by_school_size = average_scores_by_school_size.rename_axis("School Size")
average_scores_by_school_size
```
